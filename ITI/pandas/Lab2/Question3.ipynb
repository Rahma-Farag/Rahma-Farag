```python
import pandas as pd
```

```python
df = pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')
```

```python
df.head(2)
```

```python
df.set_index('customerID', inplace= True)
df.info()
```

```python
#show the duplicated records
df[df.duplicated()]
```

```python
#remove them
df.drop_duplicates(inplace = True)
df.info()
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
df.info()
```

```python
df['TotalCharges'].mean()
```

```python
filling_values = {'TotalCharges':df['TotalCharges'].mean()}
df = df.fillna(value=filling_values)
```

```python
#df.info()
```

```python
categorical_cols = ['gender', 'Partner', 'Dependents', 'PhoneService', 'MultipleLines', 
                    'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 
                    'TechSupport', 'StreamingTV', 'StreamingMovies',
                    'PaperlessBilling', 'PaymentMethod']
#pd.get_dummies(df, columns = categorical_cols)
```

```python
df = pd.get_dummies(df, columns = categorical_cols)
#df.info()
```

```python
ordinal_encoding_mapping = {"Month-to-month":1, "One year":2, "Two year":3}
df["Contract"] = df["Contract"].replace(ordinal_encoding_mapping)
#df.info()
```

```python
df['Contract'] = pd.to_numeric(df['Contract'], errors='coerce') #f â€˜coerceâ€™ invalid parsing will be set as NaN.
```

```python
df.info()
```

```python
ordinal_encoding_mapping = {"No":0, "Yes":1}
df["Churn"] = df["Churn"].replace(ordinal_encoding_mapping)
df["Churn"] = pd.to_numeric(df["Churn"], errors='coerce')
df.info()
```

```python
target_labels = df['Churn'].copy()
target_labels
```

```python
df.drop(columns = 'Churn', inplace= True)
df.info()
```

```python
df_norm=(df-df.min())/(df.max()-df.min())
df_norm.head(4)
```

```python
df_norm.to_csv("train_data.csv", index= True, sep=";")
target_labels.to_csv("train_labels.csv", index=True, sep=";")
```
