```python
from tensorflow.python.keras.models import Sequential
import tensorflow.python.keras as tf
from keras.preprocessing.image import ImageDataGenerator
```

```python
import cv2
import os
import glob 
import numpy as np
import pandas as pd
import time
import matplotlib.pyplot as plt
import zipfile as zf
import random
from datetime import datetime
from sklearn.svm import SVC
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
from sklearn.decomposition import PCA
from scipy.fftpack import dct as dct
from sklearn.feature_selection import SelectKBest,chi2
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
train_datagen = ImageDataGenerator()
training_set = train_datagen.flow_from_directory("Reduced MNIST Data/Reduced Trainging data",
                                                 target_size = (32, 32),
                                                 color_mode="grayscale",
                                                 batch_size = 200,
                                                 class_mode = "categorical")
```

```python
test_datagen = ImageDataGenerator()
test_set = test_datagen.flow_from_directory("Reduced MNIST Data/Reduced Testing data",
                                            target_size = (32, 32),
                                             color_mode="grayscale",
                                            batch_size = 200,
                                            class_mode = "categorical")
```

```python
matrix =[('CNN','n1 = 6'),('CNN','n1 = 8')]
Indices =pd.MultiIndex.from_tuples(matrix)
columns=[('Accuracy','Training Time','Testing Time')]
Columns=pd.MultiIndex.from_product(columns)
Output_Matrix =pd.DataFrame(index= Indices,columns=Columns)
Output_Matrix
```

```python
model = Sequential()

model.add(tf.layers.Conv2D(filters=6, kernel_size=(5, 5), activation='relu', input_shape=(32,32,1)))
model.add(tf.layers.AveragePooling2D(pool_size=2, strides=2))

model.add(tf.layers.Conv2D(filters=16, kernel_size=(5, 5), activation='relu'))
model.add(tf.layers.AveragePooling2D(pool_size=2, strides=2))

model.add(tf.layers.Flatten())

model.add(tf.layers.Dense(units=120, activation='relu'))

model.add(tf.layers.Dense(units=84, activation='relu'))

model.add(tf.layers.Dense(units=10, activation = 'softmax'))
```

```python
model.summary()
```

```python
model.compile(loss=tf.losses.categorical_crossentropy, optimizer='adam', metrics=['accuracy'])
```

```python
STEP_SIZE_TRAIN= training_set.n//training_set.batch_size
STEP_SIZE_TEST=test_set.n//test_set.batch_size
```

```python
start = time.time()
model.fit(training_set,
        steps_per_epoch=STEP_SIZE_TRAIN,
        epochs=100
         )
t2 = time.time() - start
print("Total time: ", t2, "seconds")
Output_Matrix.loc[('CNN','n1 = 6')]['Training Time']=t2 
```

```python
x_test = []
y_test = []
for i in range(STEP_SIZE_TEST):
    x, y = test_set.next()
    x_test.extend(x) 
    y_test.extend(y) 
x_test = np.array(x_test)
y_test = np.argmax(y_test,axis=1)
start = time.time()
pred = model.predict(x_test)
    
t2 = time.time() - start
print("Total time: ", t2, "seconds")
Output_Matrix.loc[('CNN','n1 = 6')]['Testing Time']=t2
predicted=np.argmax(pred,axis=1)
acc =  accuracy_score(y_test, predicted)
Output_Matrix.loc[('CNN','n1 = 6')]['Accuracy']= acc
```
